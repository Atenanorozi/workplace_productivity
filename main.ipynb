# Question 1:

## میکنید؟ standarize  سوال ۱: آیا داده ها را 

```python
cluster_features = [
    "noise_level_db",
    "light_exposure_lux",
    "temperature_c",
    "crowding_index",
    "meeting_hours",
    "remote_work_pct",
    "green_space_score",
    "break_area_score"
]
```

```python
scaler = StandardScaler()
scaler_feature = scaler.fit_transform(df[cluster_features])
```

استفاده میکنیم تا همه ویژگی ها در یک مقیاس قرار بگیرند Standardscaler از  (db, lux, c, index, pct,...)به دلیل تفاوت واحد اندازه گیری متغبر ها  

##  سوال ۲: روش خوشه بندی و تعداد خوشه ها؟

```python
wcss = []  # Within-Cluster Sum of Squares

for k in range(1,11):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(scaler_feature)
    wcss.append(model.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

ارنج ملاک است که میبینیم روی ۳ است پس تعداد خوشه هامون ۳ میشه elbow method در 

wcss -> Within-Cluster Sum of Squares  -> مجموع فاصله هر داده تا مرکز خوشه خودش 

```python
for k in range(2,8):
    model = KMeans(n_clusters=k, random_state=42)
    labels = model.fit_predict(scaler_feature)
    print(f"k_count: ", k, f"silhouette_score: ", silhouette_score(scaler_feature, labels))
```

 برای کیفیت خوشه بندی استفاده میشود  silhouette_score تابع  

 قابل مقایسه نیست و باید حداقل دو خوشه وجود داشته باشد silhouette_score  از ۲ شروع میشود؟ چون وقتی فقط یک خوشه داشته باشیم  range چرا 

 مشخص میکند هر داده عضو کدام خوشه است predict() مدل روی داده ها اموزش میبیند و بعد  fit() دوکار را امجام میدهد اول  fit_predict  دستور 

## سوال ۳: دلیل انتخاب خوشه؟

  تعداد ۳ خوشه انتخاب شد silhouette score  , elbow method  با استقاده از 

```python
kmeans = KMeans(n_clusters=3, random_state=42)

df['cluster_label'] = kmeans.fit_predict(scaler_feature)
```

```python
df.groupby("cluster_label")[cluster_features].mean()
```
